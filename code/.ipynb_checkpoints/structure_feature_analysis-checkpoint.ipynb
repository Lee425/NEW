#### ùë¶‚Äôs connection inclination distributions for the six networks where ùë¶ denotes the other end of eachlink.
    The dashed line corresponds to the average value of the connection inclinations in each plot. The value shown on the left side of the line is the percentage of the small weights and that shown on the right side of the line is the percentage of the large weights.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from random import sample
import glob
```

```python
def per_feature(df,arr):
    temp = df[arr].value_counts().reset_index()
    temp.columns = [arr,"count"]
    temp = temp.sort_values(by=arr)
    temp = temp.reset_index(drop=True)
    return temp
```

```python
def feature_tools(data,arr,weight_category):
    minw = data[arr].min()
    maxw = data[arr].max()
    category = (maxw - minw) / weight_category
    start = minw + category
    lt = [start]
    temp = start
    for i in range(weight_category - 1):
        temp += category
        lt.append(temp)

    s = []
    for cc in lt:
        temp_sum = data[data[arr] <= cc]['count'].values.tolist()
        s.append(temp_sum)

    pp = []
    for i in range(len(s)):
        if i == 0:
            number = sum(s[i])
            pp.append(number)
        else:
            number = sum(s[i]) - sum(s[i - 1])
            pp.append(number)

    pre_df = pd.DataFrame({"count": pp, "start_point": lt})
    pre_df['percentage'] = pre_df['count'] / pre_df['count'].sum()
    return pre_df,category
```

```python
def plot_f(new_temp,arr,per_mean,net,category):
    new_temp['start_point'] = new_temp['start_point']- category
    xx = new_temp['start_point'].values.tolist()
    yy = new_temp['percentage'].values.tolist()
    low_mean = round(new_temp[new_temp['start_point']< per_mean]['percentage'].sum(),2)
    high_mean = round((1. - low_mean),2)
    plt.xscale('log')
    plt.yscale('log')
    if net == 'celegans':
        net = 'C.elegans'
        plt.ylim((1e-4,1)) 
        plt.xlim((1,1e2))
    elif net == 'celegansneural':
        net = 'Neural network'
        plt.ylim((1e-4,1)) 
        plt.xlim((1e-1,1e2))
    elif net == 'netscience':
        net = 'Netscience'
        plt.ylim((1e-4,1)) 
        plt.xlim((1e-1,1e2))
    elif net == 'ploblogs':
        net = 'P.blogs'
        xx[0] = xx[0] + 1.
        plt.ylim((1e-4,1e1))
        plt.xlim((1e-1,1e2))
    elif net == 'ussocial':
        net = 'UC social'
        plt.ylim((1e-4,1)) 
        plt.xlim((1e-1,1e2))
    elif net == 'wiki':
        net = 'Wiki'
        plt.ylim((1e-6,1)) 
        plt.xlim((1,1e3))
    plt.scatter(xx,yy,color='blue',alpha=0.8)
    plt.title(net,fontsize=20)
    if arr == 'per1_n':
        arr = 'x\'s connection inclination'
    elif arr == 'per2_n':
        arr = 'y\'s connection inclination'
    plt.xlabel(arr,fontsize=15,verticalalignment="top")
    plt.ylabel("Percentage",fontsize=15)
    plt.vlines(per_mean,1e-8, 1,  colors = "c", linestyles = "dashed")
    plt.text(per_mean-10, yy[0], 'left='+str(low_mean), size = 15,alpha=0.8)
    plt.text(per_mean+10, yy[2], 'right='+str(high_mean), size = 15,alpha = 0.8)
    plt.show()
```

```python
def pre_entropy_diff(source_dir):
    for path in glob.glob(os.path.join(source_dir, 'wiki.csv')):
        net = os.path.basename(path).split(".")[0]
        df_original = pd.read_csv(path)
        per1_mean = df_original['per1_n'].mean()
        per2_mean = df_original['per2_n'].mean()
        weight_category = len(np.unique(df_original['weight']))
        
        df_per1 = per_feature(df_original,'per1_n')
        per1_plot,category = feature_tools(df_per1,'per1_n',weight_category)
        
        df_per2 = per_feature(df_original,'per2_n')
        per2_plot,category2 = feature_tools(df_per2,'per2_n',weight_category)
        print("------------------------{}--{}------------------------".format(net,weight_category))
        plot_f(per1_plot,'per1_n',per1_mean,net,category)
        plot_f(per2_plot,'per2_n',per2_mean,net,category2)
```

```python
pre_entropy_diff('../data')
```

### Weight distributions for the six networks.
    The dashed line corresponds to the average value of the weights in each plot. The value shown on the left side of the line is the percentage of the small weights and that shown on the right side of the line is the percentage of the large weights.
#### Take the example of the wiki web

```python
wikipedia_threads  = pd.read_csv("../data/wiki.csv")
wikiweight_mean = wikipedia_threads['weight'].mean()
minmean_weight = round(len(wikipedia_threads[wikipedia_threads['weight']<=wikiweight_mean])/len(wikipedia_threads),2)
maxmean_weight = round(1. - minmean_weight, 2)
wikipedia_threads_weight = wikipedia_threads["weight"].value_counts().reset_index()
wikipedia_threads_weight.columns = ["weight","count"]
weight_sum = wikipedia_threads_weight["count"].sum()
wikipedia_threads_weight["weight_percentage"] = wikipedia_threads_weight["count"] / weight_sum
wikipedia_threads_weight = wikipedia_threads_weight.sort_values(by="weight")
```

```python
w = wikipedia_threads_weight["weight"].values
p = wikipedia_threads_weight["weight_percentage"].values
plt.xscale('log')
plt.yscale("log")
plt.scatter(w,p,color="red")
plt.title("Wiki",fontsize=20)
plt.vlines(wikiweight_mean,1e-6, 1,  colors = "c", linestyles = "dashed")
plt.text(wikiweight_mean-2, 1e-5, "{:.0f}%".format(minmean_weight*100), size = 15,alpha=0.8)
plt.text(wikiweight_mean+2, 1e-5, "{:.0f}%".format(maxmean_weight*100), size = 15,alpha = 0.8)
plt.xlabel('Weight',fontsize=15,verticalalignment="top")
plt.ylabel("Percentage",fontsize=15)
plt.tight_layout()
plt.savefig(r"D:\netwrok\graph\wiki.png")
plt.show()
```
